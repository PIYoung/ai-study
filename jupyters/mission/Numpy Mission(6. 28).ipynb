```python
noise = np.random.normal(0.5, 3.0 ,(246522))
# [미션 3-1] 코드 해석
# 노이즈 이미지 생성
# numpy.random.normal(loc=0.0, scale=1.0, size=None)
# loc: 분포의 평균("중심")입니다.
# scale: 분포의 표준 편차(확산 또는 "너비")입니다. 음수가 아니어야 합니다.

print(noise.shape)
# print(noise[0,:2])  # 에러 요인만 파악
# 1차원 array를 2차원 array처럼 사용하려고 했음
```

```python
# [미션] noise값을 362,227,3 으로 reshape함
noise = noise.reshape(362, 227, 3)
print(noise[0, :2])
print(noise.shape)
```

```python
# 노이즈값이 추가된 결과물이 제대로 나올수 있도록 해결
plt.imshow(noise)
plt.show()

noise[noise >= 0] = 1    
noise[noise < 0] = 0

print(noise.min(), noise.max())

# casting
noise = noise.astype(np.int16)
# scaling
noise *= 255

plt.imshow(imgArr + noise)
```

```python
# tmp1=imgArr 과 tmp1=imgArr.copy() 의 차이점에 대하여 알기 # 구글검색, 파이썬 깊은복사
# tmp1이미지 -> 이미지의 모든 배열숫자값이 200이상이면 0으로 변경함
# tmp2이미지 -> 이미지의 모든 배열숫자값이 100이하이면 255로 변경함

tmp1 = imgArr.copy()
tmp2 = imgArr.copy()

print(tmp1[0, :5])

tmp1[tmp1 >= 200] = 0
tmp2[tmp2 <= 100] = 255

plt.imshow(tmp1)
```
