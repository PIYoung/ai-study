```python
## 샘플코드 '기호_1 공약.pdf'
import pdfplumber

textTotal = ''
pdf = pdfplumber.open('missions/미션3, 데이콘 대통령연설문 시각화/기호_1 공약.pdf')

for pageNum in range(len(pdf.pages)):
    textTotal = textTotal + pdf.pages[pageNum].extract_text()

print(textTotal)
```

```python
# # pdf의  엘리먼트 속성을 이용해서 작업해도 됨
# # https://bslife.tistory.com/61
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer

for page_layout in extract_pages('missions/미션3, 데이콘 대통령연설문 시각화/기호_1 공약.pdf'):
    for element in page_layout:
        if isinstance(element, LTTextContainer):
            print(element.get_text())
```

```python
#wordCloud에서 사용한 패키지
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk

from wordcloud import WordCloud
from PIL import Image
from konlpy.corpus import kolaw
from soynlp.utils import DoublespaceLineCorpus
from soynlp.noun import LRNounExtractor_v2
```

```python
with open("모든후보.txt", "w", encoding='UTF-8') as text_file:
    text_file.write(textTotal)
```

```python
corpus_fname = '모든후보.txt' #여기서 모든후보란 1~4번의 후보를 말한다.
sents = DoublespaceLineCorpus(corpus_fname, iter_sent=True)
len(sents)
```

```python
#명사 추출
noun_extractor = LRNounExtractor_v2(verbose=True, extract_compound=True)
noun_extractor.train(sents)

nouns = noun_extractor.extract()
nouns
```

```python
#mask 이미지 불러오기
im = Image.open('free-icon-mugunghwa-1285143.png')
mask_arr = np.array(im)
```

```python
#wordCloud 구현 함수
def displayWordCloud(mask,
                     colormap,
                     data=None,
                     backgroundcolor='white',
                     width=800,
                     height=600):
    wordcloud = WordCloud(font_path='C:\windows/fonts/malgun.ttf',
                          stopwords=stopwords_kr,
                          mask=mask,
                          background_color=backgroundcolor,
                          colormap=colormap,
                          width=width,
                          height=height).generate(data)
    plt.figure(figsize=(15, 10))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
```

```python
# 추출된 명사를 찍어봅니다.
%time displayWordCloud(mask_arr,None, ' '.join(nouns))
```
