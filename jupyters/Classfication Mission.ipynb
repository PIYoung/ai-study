# Y값의 Class 컬럼을 수치로 바꾸어 표현

````python
# L, M, H를 숫자로 바꾸어 표현하기 (eg. L: -1, M: 0, H:1)
# Hint) DataFrame의 map() 메소드를 사용
# 파이썬 Map함수를 이용하여 작업


def change_value(x):
    if x == 'L':
        return -1
    elif x == 'M':
        return 0
    else:
        return 1


df['Class_value'] = df['Class'].map(lambda x: change_value(x))
df
````

````python
# 숫자로 바꾼 Class_value 컬럼을 이용해 다양한 시각화 수행하기
# 차트 해석
# 성별에 따른 학생의 평균 성적 등급
# 여학생이 남학생보다 평균적으로 성적 등급이 높다

gb = df.groupby('gender').mean()['Class_value']
plt.bar(gb.index, gb)
````

````python
# 숫자로 바꾼 Class_value 컬럼을 이용해 다양한 시각화 수행하기
# 차트 해석
# 전공에 따른 학생의 평균 성적 등급

nb = df.groupby('Topic').mean()['Class_value']
nb = nb.sort_values()
plt.barh(nb.index, nb)
````

````python
# 숫자로 바꾼 Class_value 컬럼을 이용해 다양한 시각화 수행하기
# 차트 해석
# 결석일수에 따른 학생의 평균 성적 등급

ab = df.groupby('StudentAbsenceDays').mean()['Class_value']
plt.bar(ab.index, ab)
````

# 모델 학습을 위한 데이터 전처리

## get_dummies()를 이용하여 범주형 데이터 전처리하기

````python
df.columns
````

````python
# pd.get_dummies()를 이용해 범주형 데이터를 one-hot 벡터로 변환하기
# ['ParentschoolSatisfaction', 'Class', 'Class_value'] 는 drop하여서 변수에서 제외

# X, y데이터 분리함
X = df.loc[:, df.columns != 'Class_value']
y = df['Class_value']

X = X.drop(['ParentschoolSatisfaction', 'Class'], axis=1)
X = pd.get_dummies(X)
X[:1], y[:5]
````

## 학습데이터와 테스트데이터 분리하기 

- X_train, X_test, y_train, y_test 데이터셋 제작
- 테스트사이즈 0.3으로 분리/ 같은 데이터셋을 추출할수 있도록 random번호 지정

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.8,
                                                    random_state=527)

X_train[:1], X_test[:1], y_train[:5], y_test[:5]
````

# Classification 모델 학습하기

## Logistic Regression 모델 생성/학습하기

- max_iter 옵션값 조정을 통해 학습성능 높이기

````python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)
````

# 모델 학습 결과 평가하기

- Predict를 수행하고 classification_report() 결과 출력하기
- 혼돈행렬 시각화 하기

````python
from sklearn.metrics import classification_report

y_predict = model.predict(X_test)
cr = classification_report(y_test, y_predict)
cr
````

````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
````

## 참고 Logistic Regresssion 모델 계수로 상관성 파악하기

````python
model.classes_
````

````python
model.coef_.shape
````

````python
# Logistic Regression 모델의 coef_ 속성을 plot하기
fig = plt.figure(figsize=(15, 8))
plt.bar(X.columns, model.coef_[0, :])
plt.xticks(rotation=90)
plt.show()
````

````python
# Logistic Regression 모델의 coef_ 속성을 plot하기
fig = plt.figure(figsize=(15, 8))
plt.bar(X.columns, model.coef_[1, :])
plt.xticks(rotation=90)
plt.show()
````

# 의사결정트리 또는 랜덤 포레스트 모델로 학습하기

````python
from sklearn.tree import DecisionTreeClassifier

# 의사결정트리
tree = DecisionTreeClassifier(criterion='entropy',
                              max_depth=100,
                              random_state=527).fit(X_train, y_train)
````

````python
cm = confusion_matrix(y_test, tree.predict(X_test))
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
````

````python
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트
forest = RandomForestClassifier(n_estimators=100,
                                oob_score=True,
                                random_state=527).fit(X_train, y_train)
````

````python
cm = confusion_matrix(y_test, forest.predict(X_test))
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
````
