## <font color=blue> [1-2] 단위행렬(identityp matrix)
    선형대수학에서, 단위 행렬(영어: unit matrix) 또는 항등 행렬(영어: identity matrix)은 주대각선의 원소가 모두 1이며 나머지 원소는 모두 0인 정사각 행렬이다.  다른 행렬과 곱해도 그 행렬 자신이 나온다는 특징을 갖는다
    
    - 항등원(identity)에서 ‘항등’이란 ‘항상 같다’는 의미고, ‘원’은 원소를 의미함.
    - 즉, 임의의 수 A에 어떤 수를 연산한 후에도 처음의 수 A가 나오게 하는 수를 항등원이라고 함.
        - A + 0 = A라는 식이 있을 때 원소 A에 대해 실수를 더했음에도 항상 A 결과가 나온다면 0은 덧셈에 대해 항등원임
        - 또한 A×1 = A가 되므로 곱셈에서 항등원도 1임

## <font color=red> [4] 직교행렬(Orthogonal matrix)
    
   

    - 선형대수학에서, 직교 행렬(直交行列, orthogonal matrix)은 행벡터와 열벡터가 유클리드 공간의 정규 직교 기저를 이루는 실수 행렬 
    - 직교 행렬의 정의는 모든 column들이 orthonormal set을 이루는 행렬이다. 
        - orthonormal set은 orthogonal + normal의 합성어임.
        - orthogonal, 즉 모든 column 벡터들이 서로 직교한다는 뜻으로 수식적으로는 내적 (inner product)이 0을 뜻함.
        - 즉 normal, 모든 벡터의 크기가 1로 맞춰져 있다는 것임(참고: 벡터의 크기는 자기 자신과 내적한 뒤, 제곱근으로 구함)

```python
## 역행렬이 존재하지 않는 선형종속
v1 = [4, 2]
v2 = [8, 4]
V = np.array([v1, v2])

import numpy.linalg

print(V)
print(
    numpy.linalg.det(V))  #  print(4*4-2*8)      # 선형종속,   # 0으로 역행렬이 존재하지 않음.
print('rank % d' % np.linalg.matrix_rank(V)
      )  # 랭크계산, 1이면 선형종속 # https://thebook.io/080246/part03/ch10/unit19/03-18/

print('-' * 100)
## 역행렬이 존재하지 않음으로 단위행렬이 존재하지 않음.
B = np.array([v1, v2])
c = np.array([0, 0]).reshape(-1, 1)
c
Bc = np.hstack([B, c])
Bc
sp.Matrix(Bc).rref()
```

```python
## 역행렬이 존재하는 선형독립
v1 = [3, 0]
v2 = [0, -2]
V = np.array([v1, v2])
print(V)
print(numpy.linalg.det(V))  # print(3*-2-0*0)   # 0이 아님으로 역행렬 존재
print(
    'rank % d' % np.linalg.matrix_rank(V)
)  # 랭크계산, 2이상이면 선형독립 # https://thebook.io/080246/part03/ch10/unit19/03-18/

print('-' * 100)
## 역행렬이 존재함으로 단위행렬이 생성됨
B = np.array([v1, v2])
c = np.array([0, 0]).reshape(-1, 1)
c
Bc = np.hstack([B, c])
Bc
sp.Matrix(Bc).rref()
```

###  <font color=blue>  [4-3] 직교벡터
- 두 벡터 사이의 각도가 90도를 이루는 것을 직교 벡터(orthogonal vector)라고 함
- https://thebook.io/080246/part03/ch10/unit20/03/
    
||a||2 + ||b||2 = ||a + b||2

```python
### 아래 두 자료는 직교 행렬임
# https://thebook.io/080246/part03/ch10/unit20/03-03/
def orthogonal(v1, v2):
    return (np.sum(v1**2 + v2**2))**0.5 == (np.sum((v1 + v2)**2))**0.5


a = np.array([0, -1, 0])
b = np.array([1, 0, 0])
c = np.array([0, 0, 1])
d = np.array([0, 1, 1])

print(orthogonal(a, b), orthogonal(a, c), orthogonal(b, c), orthogonal(c, d))
print(np.dot(a, b), np.dot(a, c), np.dot(b, c), np.dot(c, d))

##
# np.sum(x.T*x+y.T*y)==np.sum((x+y).T*(x+y))
# np.dot(x.T,x)+np.dot(y.T,y) == np.dot((x+y).T, (x+y))
```

```python
# 코사인각도값을 이용하여서 직교행렬 여부 확인하기
# 코사인각도값이 0 이나오면 직교행렬임 (두 벡터가 이루는 각이 작을수록 유사도가 높고, 각이 클수록 유사도가 낮습니다.)
from numpy.linalg import norm


def cos_sim(A, B):
    return np.dot(A, B) / (norm(A) * norm(B)
                           )  # norm 옵션없으면 기본값 norm2임 (제곱값의합**0.5)


cos_sim(a, b), cos_sim(a, c), cos_sim(b, c), cos_sim(c, d)
```

```python
# 해당 데이터를 그대로 진행하면 너무 분산이 너무 커지기 때문에 데이터 표준화를 진행합니다.
# 표준화는 평균 0, 분산 1이 되도록 데이터 스케일을 제거하는 행위를 의미합니다.

import numpy as np
import math
import matplotlib.pyplot as plt


def scale_less(data):
    return np.round((data - np.mean(data, axis=0)) / np.std(data, axis=0), 1)


korean = scale_less([95, 90, 80, 60, 40, 80, 95, 30, 15, 60])
english = scale_less([95, 95, 75, 70, 35, 80, 90, 25, 10, 70])

print('국어성적 z-score', korean)
print('영어성적 z-score', english)

plt.plot(korean, english, 'ro')
plt.axis([-5, 5, -5, 5])
plt.grid()
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def covariance_matrix(data1, data2):
    korean = data1
    english = data2

    korean_average = sum(korean) / len(korean)
    english_avergrage = sum(english) / len(english)

    korean_deviation = [a - korean_average for a in korean]  # 편차
    english_deviation = [a - english_avergrage for a in english]  # 편차

    korean_variance = [(a - korean_average)**2
                       for a in korean]  # 분산 - 표준편차 1 - 음수를 없애기 위해 편차 제곱
    english_variance = [(a - english_avergrage)**2
                        for a in english]  # 분산 - 표준편차 1 - 음수를 없애기 위해 편차 제곱

    korean_variance = sum(korean_variance) / len(
        korean_variance)  # 분산 - 표준편차 2 - 평균을 구하여 분산을 구한다
    english_variance = sum(english_variance) / len(
        english_variance)  # 분산 - 표준편차 2 - 평균을 구하여 분산을 구한다

    korean_std_deviation = math.sqrt(
        float(korean_variance))  # 표준편차 3 - 제곱을 했기 떄문에 제곱근을 하여 표준편차를 구한다
    english_str_deviation = math.sqrt(
        float(english_variance))  # 표준편차 3 - 제곱을 했기 떄문에 제곱근을 하여 표준편차를 구한다

    data_len = len(korean_deviation) if len(korean_deviation) < len(
        english_deviation) else len(english_deviation)

    cov = [
        korean_deviation[idx] * english_deviation[idx]
        for idx in range(0, data_len)
    ]
    cov = sum(cov) / len(cov)
    relation = cov / (korean_std_deviation * english_str_deviation)

    print('국어성적 분산: %s' % (np.round(korean_variance, 2)))
    print('영어성적 분산: %s' % (np.round(english_variance, 2)))

    print('국어성적 표준편차: %s' % (math.sqrt(np.round(korean_std_deviation, 2))))
    print('영어성적 표준편차: %s' % (math.sqrt(np.round(english_str_deviation, 2))))

    print('공분산: %s' % (np.round(cov, 2)))
    print('상관계수: %s' % (relation))
    return [
        [np.round(korean_variance, 2),
         np.round(cov, 2)],  # 분산, 공분산
        [np.round(cov, 2), np.round(english_variance, 2)]  # 공분산, 분산
    ]


def scale_less(data):
    return np.round((data - np.mean(data, axis=0)) / np.std(data, axis=0), 1)


korean = scale_less([95, 90, 80, 60, 40, 80, 95, 30, 15, 60])
english = scale_less([95, 95, 75, 70, 35, 80, 90, 25, 10, 70])

print('국어성적 z-score', korean)
print('영어성적 z-score', english)

# plt.plot(korean, english, 'ro')
# plt.axis([-5, 5, -5, 5])
# plt.grid()
# plt.show()

cov_mat = covariance_matrix(korean, english)
print('공분산 행렬:', cov_mat)
```

```python
#고유값, 고유벡터 구하기

mat = np.array(cov_mat)

w, v = np.linalg.eig(mat)
print('고유값', np.round(w, 2))
print('고유벡터', np.round(v, 2))
```

```python
#데이터에 고유벡터 적용해보기

mat = np.array(cov_mat)

w, v = np.linalg.eig(mat)
print('고유값', np.round(w, 2))
print('고유백터', np.round(v, 2))

plt.plot(korean, english, 'ro')
plt.axis([-5, 5, -5, 5])

plt.quiver(0, 0, v[0][0], v[0][1], color='blue')
plt.grid()
plt.show()

# 고유값은 분산의 합과 같음 0.98+1.03, 0.02+ 1.99
```
